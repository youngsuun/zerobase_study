````python
name_list_raw = driver.find_elements_by_class_name('quickResultLstCon')
name_list_raw[1].get_attribute('data-name')
````

````python
name_list = [option.get_attribute('data-name') for option in name_list_raw]
len(name_list)
````

````python
# 매장 주소 add_list 에 정리(리스트형)
# selenium으로 했을 때 문제점 ; 보이는 화면에서만 데이터 추출 가능.
import re

add_list = []

for option in name_list_raw:
    add_list_raw = option.find_element_by_class_name('result_details').text
    add_list.append(re.split(('\n'), add_list_raw)[0])

add_list
````

````python
# Selenium -> BeautifulSoup 
from bs4 import BeautifulSoup

req=driver.page_source  
soup=BeautifulSoup(req, 'html.parser')
````

````python
# 정규표현식으로 필요 없는 문자열 지우기
a=soup.select('li.quickResultLstCon')[0].find('p', class_='result_details').text
num = re.compile(r'\d+\-\d+')
num.sub('',a)
````

````python
import re

add_list = []
add_list_raw = soup.select('li.quickResultLstCon')

for option in add_list_raw:
    add_list_raw = option.find('p', class_='result_details').text
    num = re.compile(r'\d+\-\d+')
    add_list.append(num.sub('', add_list_raw))

add_list, len(add_list)
````

````python
# pandas로 데이터 정리

import pandas as pd

starbucks = pd.DataFrame({
    'name' : name_list,
    'address' : add_list 
})

starbucks.head()
````

## 2. 서울시 이디야커피 매장 주소 가져오기

- 진행 단계 
```
- selenium으로 홈페이지 메인 > 매장찾기 > 주소 까지 접근
- selenium으로 서울시 구 이름 입력해(send_keys) 검색하기
- BeautifulSoup으로 매장이름, 매장주소 가져오기
- pandas로 데이터 정리

- 목표데이터
    - 서울시 구
    - 이디야커피 매장 이름
    - 이디야커피 매장 주소

````python
from selenium import webdriver
````

````python
# 페이지 접근
url = 'https://www.ediya.com/contents/find_store.html'
driver = webdriver.Chrome('../driver/Chromedriver.exe')
driver.get(url)
````

````python
driver.close()
````

````python
# 주소 클릭
driver.find_element_by_xpath('//*[@id="contentWrap"]/div[3]/div/div[1]/ul/li[2]/a').click()
````

````python
# 뷰티풀 숩 임포트
from bs4 import BeautifulSoup
````

````python
# 구 입력(for문) -> 매장 이름, 주소 데이터 추출 -> 리스트 담기

name_list = []
add_list = []

import time 

for option in seoul_gu_list:
    search_bar = driver.find_element_by_id('keyword')
    search_bar.send_keys(option)
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="keyword_div"]/form/button').click()
    time.sleep(2)
    req=driver.page_source  
    soup=BeautifulSoup(req, 'html.parser')
    time.sleep(2)
    tag = soup.select('li.item')
    time.sleep(2)
    

    for i in range(len(tag)):
        raw_data = tag[i].find('dl').text
        name = raw_data.split()[0]
        name_list.extend([name])
        add_raw = raw_data.replace(name, '')  # 매점 이름 삭제해서 주소만 남기기
        add_list.extend([add_raw.strip()])  # 공백 제거

    search_bar.clear()  # 검색창 지워주기
````

````python
len(name_list), len(add_list)
````

````python
# pandas로 데이터 정리

import pandas as pd

ediya = pd.DataFrame({
    'name' : name_list,
    'address' : add_list 
})

ediya.head()
````

````python
gu_raw = []

for item in ediya.address:
    gu_raw.append(item.split()[1])

ediya['구']=gu_raw
ediya.head()
````

## 미션 : 이디야 커피는 스타벅스 커피 주변에 위치해 있는가? 

### 분석 결과 : 이디야 커피는 스타벅스 매장 근처에 전략적으로 매장을 오픈한다고 볼 수 없다.

```
분석 근거 : 
1. 스타벅스, 이디야 커피의 각 구별 매장 수를 정렬하여 비교한다.
스타벅스 매장 수가 가장 많은 곳은 강남구(87개), 가장 적은 곳은 도봉구(3개)로 구별 매장 수의 편차가 큰 편이다.
이디야 매장 수가 가장 많은 곳은 강서구(47개), 가장 적은 곳은 용산구(15개)로 구별 매장 수의 편차가 적은 편이다.
-> 구별 매장 수를 정렬하여 비교한 결과, 스타벅스와 이디야가 타켓팅하는 구, 지역이 각자 다르다는 것을 알 수 있다.

2. 스타벅스, 이디야 커피의 서울시 구별 매장 수를 정규화한다.
(서울시 각 구에 분포되어 있는 매장 비율을 비교하기 위함)
-> folium으로 시각화한 결과, 스타벅스와 이디야 커피의 각 구별 매장 비율이 다르다는 것을 알 수 있다.
-> 그나마 비슷한 색상을 띄고 있는 '강남구'에서 각 매장의 위치를 마커로 표시하여, 스타벅스 매장과 이디야 커피 매장이
직관적으로 봤을 때 서로 근처에 있는지 살펴 본다.

3. 매장 비율이 비슷한 '강남구'를 비교한다.
-> 매장 비율은 스타벅스 1, 이디야 0.95로 비슷하지만,
스타벅스 매장 수 87개 이디야 매장 수 45개로, 스타벅스 매장 수가 거의 2배나 많아,
상대적으로 이디야가 위치한 곳에 스타벅스가 위치할 확률보다,
스타벅스가 위치한 곳에 이디야가 위치해 있을 확률이 더 크다는 점을 인지해야 한다.
-> folium 마커로 시각화한 결과, 스타벅스가 주변에 없는 이디야 매장의 개수는 12개로, 약 27% 이다.
-> 매장 수의 차이가 약 2배로 , 분석 근거를 활용하는데 적합하지 않다.

4. 매장 수가 비슷한 '영등포구'를 비교한다.
-> 40,39개로 매장 수가 비슷한 '영등포구'에 folium 마커로 시각화한 결과,
스타벅스는 여의도에 몰려있고, 이디야는 영등포동(영등포구의 남쪽)에 몰려있는 것을 알 수 있다.
```
